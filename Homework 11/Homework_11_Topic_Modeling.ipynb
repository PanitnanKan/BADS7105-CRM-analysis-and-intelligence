```python
import pandas as pd
import pythainlp
from pythainlp.corpus import thai_stopwords
from pythainlp.tokenize import word_tokenize
from pythainlp.util import normalize
```

```python
df = pd.read_csv('CustomerReviews.csv')
```

```python
df.head()
```

###Tokenize

```python
removed_words = [' ','  ','\r','\n','ร้าน','(',')',':','/','-','+','!','.','?','++','::','']
stopwords = list(pythainlp.corpus.thai_stopwords())
del_words = removed_words + stopwords
def tokenize_sentence(sentence):
    text = normalize(sentence)
    list_word = word_tokenize(text)
    list_word_clean = [i for i in list_word if i not in del_words]
    word_clean = ','.join(list_word_clean)
    return word_clean
    
df['word_clean'] = df['Review'].apply(lambda x: tokenize_sentence(x))
df.head()
```

###Dictionary

```python
import gensim
import gensim.corpora as corpora
from gensim.models import CoherenceModel
```

```python
data = df['word_clean'].to_list()
docs = []
for i in data:
  if len(i) > 0:
    docs.append(i)

texts=[[text for text in doc.split(',')] for doc in docs]
dictionary = gensim.corpora.Dictionary(texts)
print(dictionary.token2id.keys())
```

```python
corpus = [dictionary.doc2bow(text,allow_update=True) for text in texts]
print(corpus[:1])
```

```python
word_freq=[[(dictionary[id],frequence) for id,frequence in couple] for couple in corpus]
print(word_freq[:1])
```

###Topic Modelling

```python
from pprint import pprint
# number of topics
num_topics = 4
id2word=dictionary.id2token
# Build LDA model
#lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           #id2word=id2word,
                                           #num_topics=num_topics, 
                                           #random_state=100,
                                           #update_every=1,
                                           #chunksize=100,
                                           #passes=10,
                                           #alpha='auto',
                                           #per_word_topics=True)

# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics)

# Print the Keyword in the 4 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
